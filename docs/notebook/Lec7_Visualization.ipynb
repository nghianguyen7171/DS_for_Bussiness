<hr>

**üìê ƒêi·ªÅu ch·ªânh kho·∫£ng c√°ch gi·ªØa c√°c subplot:**

- Kho·∫£ng c√°ch gi·ªØa c√°c subplot c≈©ng ƒë∆∞·ª£c ch·ªâ ƒë·ªãnh
- ƒê∆∞·ª£c ch·ªâ ƒë·ªãnh t∆∞∆°ng ƒë·ªëi v·ªõi chi·ªÅu cao v√† chi·ªÅu r·ªông c·ªßa ƒë·ªì th·ªã

## Ch√∫ th√≠ch v·∫Ω tr√™n m·ªôt ƒë·ªì th·ªã con
<hr>

**‚úèÔ∏è Ch√∫ th√≠ch v√† vƒÉn b·∫£n tr√™n ƒë·ªì th·ªã:**
- V·∫Ω c√°c ch√∫ th√≠ch ƒë·ªì th·ªã t√πy ch·ªânh: vƒÉn b·∫£n, m≈©i t√™n ho·∫∑c c√°c h√¨nh d·∫°ng kh√°c
- S·ª≠ d·ª•ng c√°c h√†m `text`, `arrow`, v√† `annotate`

**üìù H√†m text:**
- `ax.text(x, y, "Hello world!", family="monospace", fontsize=10)`
- V·∫Ω vƒÉn b·∫£n t·∫°i c√°c t·ªça ƒë·ªô (x, y) ƒë√£ cho tr√™n ƒë·ªì th·ªã v·ªõi ki·ªÉu t√πy ch·ªânh

**üîó H√†m annotate:**
- Ch√∫ th√≠ch c√≥ th·ªÉ bao g·ªìm c·∫£ vƒÉn b·∫£n v√† m≈©i t√™n ch·ªâ v√†o v·ªã tr√≠ c·ª• th·ªÉ

```python
from datetime import datetime
# Create spx.csv file for the example
spx_csv_content = "date,SPX\n2003-01-02,909.03\n2007-05-04,1499.26\n2007-10-11,1554.41\n2008-03-12,1290.35\n2008-09-15,1172.22\n2010-12-31,1257.64\n2011-01-03,1271.87"
with open("spx.csv", "w") as f:
    f.write(spx_csv_content)

fig, ax = plt.subplots(figsize=(12, 6)) # TƒÉng k√≠ch th∆∞·ªõc ƒë·ªÉ d·ªÖ nh√¨n h∆°n
data = pd.read_csv("spx.csv", index_col=0, parse_dates=True)
spx = data["SPX"]

spx.plot(ax=ax, color="black")

crisis_data = [
    (datetime(2007, 10, 11), "Peak of bull market"),
    (datetime(2008, 3, 12), "Bear Stearns Fails"),
    (datetime(2008, 9, 15), "Lehman Bankruptcy")
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor="black", headwidth=4, width=2,
                                headlength=4),
                horizontalalignment="left",
                verticalalignment="top")

# Zoom in on 2007-2010
ax.set_xlim(["1/1/2007", "1/1/2011"])
ax.set_ylim([600, 1800])
ax.set_title("Important dates in the 2008‚Äì2009 financial crisis")
plt.savefig("chart13.png")
```

**üìç ƒêi·ªÉm quan tr·ªçng trong ch√∫ th√≠ch:**
- `ax.annotate`: v·∫Ω c√°c nh√£n t·∫°i c√°c t·ªça ƒë·ªô x v√† y ƒë∆∞·ª£c ch·ªâ ƒë·ªãnh
- `set_xlim` v√† `set_ylim`: ƒë·∫∑t th·ªß c√¥ng c√°c ranh gi·ªõi b·∫Øt ƒë·∫ßu v√† k·∫øt th√∫c cho ƒë·ªì th·ªã
- Thay v√¨ s·ª≠ d·ª•ng m·∫∑c ƒë·ªãnh c·ªßa matplotlib
- `ax.set_title`: th√™m m·ªôt ti√™u ƒë·ªÅ ch√≠nh cho ƒë·ªì th·ªã

**üìö T√†i li·ªáu:**
Xem th∆∞ vi·ªán matplotlib ƒë·ªÉ bi·∫øt th√™m c√°c t√πy ch·ªçn ƒë·ªãnh d·∫°ng vƒÉn b·∫£n chi ti·∫øt

**üî∑ V·∫Ω c√°c h√¨nh d·∫°ng:**
- Vi·ªác v·∫Ω c√°c h√¨nh d·∫°ng ƒë√≤i h·ªèi s·ª± c·∫©n th·∫≠n v√† t·ªâ m·ªâ h∆°n
- `matplotlib` c√≥ c√°c ƒë·ªëi t∆∞·ª£ng ƒë·∫°i di·ªán cho nhi·ªÅu h√¨nh d·∫°ng ph·ªï bi·∫øn, g·ªçi l√† `patches`
- M·ªôt s·ªë nh∆∞ `Rectangle` v√† `Circle` ƒë∆∞·ª£c t√¨m th·∫•y trong `matplotlib.pyplot`
- Th√¥ng tin ƒë·∫ßy ƒë·ªß n·∫±m trong `matplotlib.patches`

**‚úÖ C√°ch s·ª≠ d·ª•ng:**
- T·∫°o ƒë·ªëi t∆∞·ª£ng `patches` v√† th√™m n√≥ v√†o m·ªôt ƒë·ªì th·ªã con `ax` b·∫±ng ph∆∞∆°ng th·ª©c `add_patch`

# V·∫Ω ƒë·ªì th·ªã v·ªõi pandas v√† seaborn
<hr>

`matplotlib` l√† m·ªôt c√¥ng c·ª• tr·ª±c quan h√≥a d·ªØ li·ªáu ·ªü m·ª©c ƒë·ªô kh√° c∆° b·∫£n. Gi·ªëng nh∆∞ b·∫°n ƒëang *l·∫Øp r√°p* m·ªôt ƒë·ªì th·ªã t·ª´ c√°c th√†nh ph·∫ßn c∆° b·∫£n c·ªßa n√≥ bao g·ªìm: hi·ªÉn th·ªã d·ªØ li·ªáu, ch√∫ gi·∫£i, ti√™u ƒë·ªÅ, nh√£n v·∫°ch chia v√† c√°c ch√∫ th√≠ch kh√°c.

Trong `pandas`, ch√∫ng ta c√≥ th·ªÉ c√≥ nhi·ªÅu c·ªôt d·ªØ li·ªáu, c√πng v·ªõi c√°c ch·ªâ s·ªë t√™n h√†ng v√† t√™n c·ªôt. B·∫£n th√¢n `pandas` c√≥ c√°c ph∆∞∆°ng th·ª©c t√≠ch h·ª£p s·∫µn gi√∫p ƒë∆°n gi·∫£n h√≥a vi·ªác t·∫°o h√¨nh tr·ª±c quan t·ª´ c√°c ƒë·ªëi t∆∞·ª£ng `DataFrame` v√† `Series`. M·ªôt th∆∞ vi·ªán kh√°c l√† `seaborn`, m·ªôt th∆∞ vi·ªán ƒë·ªì h·ªça th·ªëng k√™ c·∫•p cao ƒë∆∞·ª£c x√¢y d·ª±ng tr√™n `matplotlib`. ∆Øu ƒëi·ªÉm c·ªßa seaborn l√† ƒë∆°n gi·∫£n h√≥a vi·ªác t·∫°o nhi·ªÅu lo·∫°i h√¨nh tr·ª±c quan ph·ªï bi·∫øn.

```python
import warnings
warnings.filterwarnings('ignore')

# B·∫£o ƒë·∫£m hi·ªÉn th·ªã inline an to√†n tr√™n nhi·ªÅu m√¥i tr∆∞·ªùng
try:
    get_ipython().run_line_magic('matplotlib', 'inline')
except Exception:
    pass

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# T·∫°o DataFrame t·ªïng h·ª£p m√¥ ph·ªèng doanh thu theo th√°ng cho 2 s·∫£n ph·∫©m
np.random.seed(0)
months = pd.date_range('2024-01-01', periods=12, freq='M')
df_demo = pd.DataFrame({
    'Month': months,
    'Revenue_A': np.random.randint(80, 140, size=12).cumsum(),
    'Revenue_B': np.random.randint(60, 120, size=12).cumsum()
})
df_demo['Total'] = df_demo['Revenue_A'] + df_demo['Revenue_B']

print('‚úÖ T·∫°o DataFrame t·ªïng h·ª£p th√†nh c√¥ng!')
df_demo.head()

# 1) pandas.plot: v·∫Ω ƒë∆∞·ªùng th·ªùi gian cho hai s·∫£n ph·∫©m
ax = df_demo.set_index('Month')[['Revenue_A', 'Revenue_B']].plot(
    figsize=(8, 4), linewidth=2, marker='o', title='Doanh thu theo th√°ng (pandas.plot)'
)
ax.set_xlabel('Th√°ng')
ax.set_ylabel('Doanh thu (ƒëv gi·∫£ l·∫≠p)')
plt.tight_layout()
plt.savefig('chart_pandas_demo.png')
plt.show()

# 2) seaborn: chuy·ªÉn sang d·∫°ng long v√† v·∫Ω barplot theo th√°ng
long_df = df_demo.melt(id_vars='Month', value_vars=['Revenue_A', 'Revenue_B'],
                       var_name='Product', value_name='Revenue')

plt.figure(figsize=(9, 4))
sns.barplot(data=long_df, x='Month', y='Revenue', hue='Product', alpha=0.9)
plt.title('So s√°nh doanh thu theo th√°ng (seaborn.barplot)')
plt.xlabel('Th√°ng')
plt.ylabel('Doanh thu (ƒëv gi·∫£ l·∫≠p)')
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig('chart_seaborn_bar_demo.png')
plt.show()

# 3) seaborn.lineplot: ƒë∆∞·ªùng xu h∆∞·ªõng t·ªïng doanh thu
plt.figure(figsize=(8, 4))
sns.lineplot(data=df_demo, x='Month', y='Total', marker='o', linewidth=2, color='#2563eb')
plt.title('Xu h∆∞·ªõng t·ªïng doanh thu (seaborn.lineplot)')
plt.xlabel('Th√°ng')
plt.ylabel('T·ªïng doanh thu (ƒëv gi·∫£ l·∫≠p)')
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig('chart_seaborn_line_demo.png')
plt.show()
```

**L∆∞u √Ω**

M·ªôt c√¥ng th·ª©c h·ªØu √≠ch cho c√°c ƒë·ªì th·ªã c·ªôt l√† tr·ª±c quan h√≥a t·∫ßn su·∫•t gi√° tr·ªã c·ªßa m·ªôt `Series` b·∫±ng c√°ch s·ª≠ d·ª•ng `value_counts`: `s.value_counts().plot.bar()`.

H√£y xem m·ªôt v√≠ d·ª• v·ªÅ t·∫≠p d·ªØ li·ªáu ti·ªÅn boa nh√† h√†ng. Gi·∫£ s·ª≠ ch√∫ng ta mu·ªën t·∫°o m·ªôt ƒë·ªì th·ªã c·ªôt x·∫øp ch·ªìng hi·ªÉn th·ªã t·ª∑ l·ªá ph·∫ßn trƒÉm c√°c ƒëi·ªÉm d·ªØ li·ªáu cho m·ªói quy m√¥ b·ªØa ti·ªác cho m·ªói ng√†y. Ch√∫ng ta t·∫£i d·ªØ li·ªáu b·∫±ng `read_csv` v√† t·∫°o m·ªôt b·∫£ng ch√©o theo ng√†y v√† quy m√¥ b·ªØa ti·ªác. H√†m `pandas.crosstab` l√† m·ªôt c√°ch thu·∫≠n ti·ªán ƒë·ªÉ t√≠nh to√°n m·ªôt b·∫£ng t·∫ßn su·∫•t ƒë∆°n gi·∫£n t·ª´ hai c·ªôt DataFrame:

```python
# D·ªØ li·ªáu m√¥ ph·ªèng th·ª±c t·∫ø v·ªõi c√°c m·∫´u v·ªÅ tip percentage theo ng√†y v√† th·ªùi gian
days = ['Thur', 'Fri', 'Sat', 'Sun']
times = ['Lunch', 'Dinner']
sexes = ['Male', 'Female']
smokers = ['Yes', 'No']

# T·∫°o ~200 observations ƒë·ªÉ c√≥ ƒë·ªß d·ªØ li·ªáu cho c√°c bi·ªÉu ƒë·ªì
tips_data = []
np.random.seed(42)  # ƒê·ªÉ c√≥ th·ªÉ t√°i l·∫≠p k·∫øt qu·∫£

for _ in range(200):
    # Ch·ªçn ng√†y (cu·ªëi tu·∫ßn c√≥ nhi·ªÅu kh√°ch h∆°n)
    day = np.random.choice(days, p=[0.15, 0.15, 0.35, 0.35])  # Sat, Sun nhi·ªÅu h∆°n

    # Ch·ªçn th·ªùi gian (d·ª±a v√†o ng√†y)
    if day in ['Sat', 'Sun']:
        time = np.random.choice(times, p=[0.1, 0.9])  # Cu·ªëi tu·∫ßn ch·ªß y·∫øu Dinner
    else:
        time = np.random.choice(times, p=[0.6, 0.4])  # Ng√†y th∆∞·ªùng c√≥ c·∫£ Lunch

    # Quy m√¥ b·ªØa ti·ªác (size) - t·ª´ 1 ƒë·∫øn 6 ng∆∞·ªùi
    size = np.random.choice([1, 2, 3, 4, 5, 6], p=[0.05, 0.35, 0.25, 0.20, 0.10, 0.05])

    # T·ªïng h√≥a ƒë∆°n - ph·ª• thu·ªôc v√†o size v√† time
    base_bill = 15 + size * 8  # Bill c∆° b·∫£n
    if time == 'Dinner':
        base_bill *= 1.3  # Dinner th∆∞·ªùng ƒë·∫Øt h∆°n
    if day in ['Sat', 'Sun']:
        base_bill *= 1.1  # Cu·ªëi tu·∫ßn c√≥ th·ªÉ ƒë·∫Øt h∆°n
    total_bill = np.random.normal(base_bill, base_bill * 0.3)
    total_bill = max(5, total_bill)  # ƒê·∫£m b·∫£o t·ªëi thi·ªÉu $5

    # T√≠nh tip - ph·ª• thu·ªôc v√†o nhi·ªÅu y·∫øu t·ªë
    # Tip percentage th∆∞·ªùng 15-20%, nh∆∞ng c√≥ th·ªÉ thay ƒë·ªïi
    base_tip_pct = 0.17
    if day in ['Sat', 'Sun']:
        base_tip_pct += 0.02  # Cu·ªëi tu·∫ßn tip t·ªët h∆°n
    if size >= 4:
        base_tip_pct -= 0.01  # Nh√≥m l·ªõn tip √≠t h∆°n m·ªôt ch√∫t
    tip_pct = np.random.normal(base_tip_pct, 0.05)
    tip_pct = max(0.05, min(0.30, tip_pct))  # Gi·ªõi h·∫°n t·ª´ 5% ƒë·∫øn 30%

    tip = total_bill * tip_pct

    # C√°c thu·ªôc t√≠nh kh√°c
    sex = np.random.choice(sexes)
    smoker = np.random.choice(smokers, p=[0.38, 0.62])  # Kho·∫£ng 38% h√∫t thu·ªëc

    tips_data.append({
        'total_bill': total_bill,
        'tip': tip,
        'sex': sex,
        'smoker': smoker,
        'day': day,
        'time': time,
        'size': size
    })

tips = pd.DataFrame(tips_data)

# T√≠nh tip_pct (tip percentage) - s·∫Ω ƒë∆∞·ª£c d√πng trong c√°c v√≠ d·ª•
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])

print(f"\n‚úÖ Tips DataFrame ƒë√£ ƒë∆∞·ª£c t·∫°o th√†nh c√¥ng!")
print(f"   - S·ªë h√†ng: {len(tips)}")
print(f"   - C√°c c·ªôt: {', '.join(tips.columns.tolist())}")
print(f"\nüìã 5 h√†ng ƒë·∫ßu ti√™n c·ªßa tips DataFrame:")
print(tips.head())
```

```python
party_counts = pd.crosstab(tips["day"], tips["size"])
party_counts = party_counts.reindex(index=["Thur", "Fri", "Sat", "Sun"])
party_counts
```

V√¨ c√≥ kh√¥ng nhi·ªÅu b·ªØa ti·ªác m·ªôt v√† s√°u ng∆∞·ªùi, ch√∫ng ta lo·∫°i b·ªè c√°c gi√° tr·ªã n√†y:

```python
# Ti·∫øp t·ª•c t·ª´ cell tr∆∞·ªõc
party_counts = party_counts.loc[:, 2:5]
party_counts
```

V·ªõi d·ªØ li·ªáu y√™u c·∫ßu t·ªïng h·ª£p ho·∫∑c t√≥m t·∫Øt tr∆∞·ªõc khi t·∫°o ƒë·ªì th·ªã, vi·ªác s·ª≠ d·ª•ng th∆∞ vi·ªán `seaborn` c√≥ th·ªÉ l√†m cho m·ªçi th·ª© ƒë∆°n gi·∫£n h∆°n nhi·ªÅu. H√£y xem x√©t t·ª∑ l·ªá ph·∫ßn trƒÉm ti·ªÅn boa theo ng√†y v·ªõi `seaborn` nh∆∞ sau:

B·ªüi v√¨ c√≥ nhi·ªÅu quan s√°t cho m·ªói gi√° tr·ªã trong `day`, c√°c thanh l√† gi√° tr·ªã trung b√¨nh c·ªßa `tip_pct`. C√°c ƒë∆∞·ªùng m√†u ƒëen ƒë∆∞·ª£c v·∫Ω tr√™n c√°c thanh bi·ªÉu th·ªã kho·∫£ng tin c·∫≠y 95% c·ªßa gi√° tr·ªã trung b√¨nh.

S·ª≠ d·ª•ng d·ªØ li·ªáu ti·ªÅn boa t·ª´ ph·∫ßn tr∆∞·ªõc, ch√∫ng ta c√≥ th·ªÉ t·∫°o bi·ªÉu ƒë·ªì t·∫ßn su·∫•t t·ª∑ l·ªá ph·∫ßn trƒÉm ti·ªÅn boa c·ªßa t·ªïng h√≥a ƒë∆°n b·∫±ng ph∆∞∆°ng th·ª©c `plot.hist` tr√™n Series:

Quy tr√¨nh th√¥ng th∆∞·ªùng l√† x·∫•p x·ªâ ph√¢n ph·ªëi n√†y d∆∞·ªõi d·∫°ng h·ªón h·ª£p (mixed) c·ªßa c√°c kernel - l√† c√°c ph√¢n ph·ªëi ƒë∆°n gi·∫£n h∆°n nh∆∞ ph√¢n ph·ªëi chu·∫©n. Do ƒë√≥, c√°c ƒë·ªì th·ªã m·∫≠t ƒë·ªô c√≤n ƒë∆∞·ª£c g·ªçi l√† ƒë·ªì th·ªã ∆∞·ªõc t√≠nh m·∫≠t ƒë·ªô theo kernel (KDE).
