# Phần 1: Thư viện NumPy

## Bài 1.1 - Tạo ma trận, đổi kiểu dữ liệu, broadcasting

```python
import numpy as np

# 1) Tạo ma trận A
A = np.array([
    [5.1, 8.7, 3.3],
    [4.4, 0.9, 9.5],
    [6.2, 7.5, 1.8]
])
print("A:\n", A)

# 2) Đổi kiểu sang số nguyên -> B
B = A.astype(int)
print("\nB (int):\n", B)
print("dtype của B:", B.dtype)

# 3) Cộng với vector float64 u (broadcast theo hàng)
u = np.array([0.5, 1.5, 2.5], dtype=np.float64)
C = B + u
print("\nC = B + u:\n", C)
print("dtype của C:", C.dtype)

# 4) Cộng với vector int32 v
v = np.array([10, 20, 30], dtype=np.int32)
D = B + v
print("\nD = B + v:\n", D)
print("dtype của D:", D.dtype)

print("\nGiải thích:")
print("- astype(int) cắt phần thập phân để tạo B kiểu int.")
print("- Khi cộng ma trận 3x3 với vector 3 phần tử, NumPy broadcasting theo từng cột.")
print("- C có dtype float vì cộng với vector float64.")
```

## Bài 1.2 - Slicing, fancy indexing, boolean indexing

```python
import numpy as np

# 1) Tạo ma trận M 6x6 chứa số từ 0 đến 35
M = np.arange(36).reshape(6, 6)
print("M:\n", M)

# 2) Trích xuất vùng trung tâm 4x4 (hàng 1:5, cột 1:5)
P = M[1:5, 1:5]
print("\nP (vùng trung tâm 4x4):\n", P)

# 3) Chọn các hàng 0 và 3 của P
selected_rows = P[[0, 3], :]
print("\nHàng 0 và 3 của P:\n", selected_rows)

# 4) Gán -99 cho cột 1 và 2 của M
M[:, [1, 2]] = -99
print("\nM sau khi gán -99 vào cột 1 và 2:\n", M)

# 5) Lấy các cột lẻ (1,3,5) rồi lọc giá trị > 20
odd_cols = M[:, [1, 3, 5]]
greater_than_20 = odd_cols[odd_cols > 20]
print("\nCác cột lẻ (1,3,5):\n", odd_cols)
print("\nGiá trị > 20 trong các cột lẻ:", greater_than_20)

print("\nGiải thích:")
print("- [1:5] lấy từ chỉ số 1 đến 4.")
print("- [[0,3], :] là fancy indexing để chọn nhiều hàng không liên tiếp.")
print("- Boolean indexing giúp lọc nhanh theo điều kiện > 20.")
```

## Bài 1.3 - Ma trận ngẫu nhiên, chuyển vị, thống kê theo hàng

```python
import numpy as np

np.random.seed(42)  # Đặt seed để kết quả có thể lặp lại

# 1) Ma trận ngẫu nhiên 4x5 trong [10, 100)
data = np.random.randint(10, 100, size=(4, 5))
print("data:\n", data)

# 2) Chuyển vị
data_T = data.T
print("\ndata.T:\n", data_T)

# 3) Max và min theo từng hàng (axis=1)
row_max = data.max(axis=1)
row_min = data.min(axis=1)
print("\nMax từng hàng:", row_max)
print("Min từng hàng:", row_min)

# 4) Chỉ số cột có giá trị lớn nhất ở mỗi hàng
argmax_col = data.argmax(axis=1)
print("\nChỉ số cột max mỗi hàng:", argmax_col)

print("\nGiải thích:")
print("- axis=1 nghĩa là tính theo từng hàng.")
print("- argmax(axis=1) trả về vị trí cột của phần tử lớn nhất trong mỗi hàng.")
```

## Bài 1.4 - Ứng dụng ma trận chi phí trong sản xuất

```python
import numpy as np

# 1) Ma trận chi phí theo sản phẩm (A,B,C) x loại chi phí (NVL, nhân công, vận hành)
costs = np.array([
    [10, 5, 2],
    [12, 7, 3],
    [9, 4, 2.5]
])
print("Ma trận costs:\n", costs)

# 2) Vector số lượng sản xuất
production_volume = np.array([1000, 500, 1500])
print("\nSố lượng sản xuất:", production_volume)

# 3) Tổng chi phí cho từng loại chi phí
# Nhân từng hàng với sản lượng tương ứng rồi cộng theo trục sản phẩm
total_cost_by_type = (costs * production_volume.reshape(-1, 1)).sum(axis=0)
print("\nTổng chi phí theo loại [NVL, Nhân công, Vận hành]:", total_cost_by_type)

# 4) Tổng chi phí cho từng sản phẩm trên 1 đơn vị sản phẩm
unit_cost_per_product = costs.sum(axis=1)
print("\nTổng chi phí/đơn vị theo sản phẩm [A,B,C]:", unit_cost_per_product)

print("\nGiải thích:")
print("- reshape(-1,1) để production_volume có dạng cột và broadcast đúng với costs.")
print("- sum(axis=0): cộng theo chiều sản phẩm để ra tổng theo từng loại chi phí.")
```

## Bài 1.5 - Hàm Sigmoid bằng NumPy

```python
import numpy as np

# 1) Tạo mảng x từ -3 đến 3, bước 0.5
x = np.arange(-3, 3.5, 0.5)
print("x:", np.round(x, 2))

# 2) Tính Sigmoid: 1 / (1 + exp(-x))
sigmoid = 1 / (1 + np.exp(-x))
print("Sigmoid(x):", np.round(sigmoid, 4))

print("\nGiải thích:")
print("- Với x âm lớn, sigmoid gần 0; với x dương lớn, sigmoid gần 1.")
print("- Đây là hàm phổ biến trong học máy để nén giá trị về khoảng (0,1).")
```

## Bài 1.6 - Định thức, khả nghịch, nghiệm hệ phương trình

```python
import numpy as np

# Chọn một ma trận khả nghịch (det != 0)
A = np.array([
    [2, 1, 1],
    [1, 3, 2],
    [1, 0, 0]
], dtype=float)
b = np.array([4, 5, 6], dtype=float)

print("A:\n", A)
print("b:", b)

# 2) Định thức và kiểm tra khả nghịch
det_A = np.linalg.det(A)
print("\nĐịnh thức det(A) =", det_A)
if abs(det_A) > 1e-10:
    print("A khả nghịch")
else:
    print("A không khả nghịch")

# 3) Ma trận nghịch đảo
A_inv = np.linalg.inv(A)
print("\nA^-1:\n", A_inv)

# 4) Nghiệm x = A^-1 b
x = A_inv @ b
print("\nNghiệm x:", x)

# 5) Kiểm tra Ax có bằng b không
check_b = A @ x
print("\nA @ x =", check_b)
print("So sánh với b:", np.allclose(check_b, b))

print("\nGiải thích:")
print("- Nếu det(A)=0 thì không thể tính nghịch đảo.")
print("- np.allclose dùng để kiểm tra gần bằng do sai số số học dấu chấm động.")
```

## Bài 2.1 - Series từ dictionary và vector hóa

```python
import pandas as pd

# 1) Tạo Series
population_dict = {
    "Hanoi": 8.5,
    "HCMC": 9.3,
    "Danang": 1.2,
    "Haiphong": 2.1
}
city_pop = pd.Series(population_dict, name="Dân số (triệu)")
print("Dân số các thành phố (triệu người):")
print(city_pop)

# 2) Trích xuất dữ liệu
print("\nDân số Hà Nội:", city_pop["Hanoi"], "triệu")
print("Dân số Đà Nẵng:", city_pop["Danang"], "triệu")

# 3) Vector hóa: đổi từ triệu người sang người
city_pop_people = (city_pop * 1_000_000).astype(int)
print("\nDân số quy đổi sang người:")
print(city_pop_people)

print("\nGiải thích: Nhân trực tiếp cả Series là phép toán vector hóa, không cần vòng lặp.")
```

## Bài 2.2 - DataFrame, thêm cột theo điều kiện, xóa cột

```python
import pandas as pd
import numpy as np

# 1) Tạo DataFrame
sales_data = {
    "nhan_vien": ["An", "Binh", "Chi", "Dung"],
    "khu_vuc": ["Bắc", "Nam", "Trung", "Bắc"],
    "doanh_so": [120, 250, 180, 90]
}
sales_df = pd.DataFrame(sales_data)
print("DataFrame ban đầu:\n", sales_df)

# 2) Cột thuong: True nếu doanh_so > 100
sales_df["thuong"] = sales_df["doanh_so"] > 100

# 3) Cột tien_thuong: 10% doanh số nếu được thưởng, ngược lại 0
sales_df["tien_thuong"] = np.where(sales_df["thuong"], sales_df["doanh_so"] * 0.1, 0)

# 4) Xóa cột khu_vuc
sales_df = sales_df.drop(columns=["khu_vuc"])

print("\nDataFrame sau khi xử lý:\n", sales_df)

print("\nGiải thích:")
print("- np.where giúp tạo cột theo điều kiện nhanh và rõ ràng.")
print("- drop(columns=[...]) dùng để xóa cột không còn cần thiết.")
```

## Bài 2.3 - Truy cập dữ liệu với loc và iloc

```python
import pandas as pd
import numpy as np

# 1) Tạo DataFrame 4x4 từ 0 đến 15
df = pd.DataFrame(
    np.arange(16).reshape(4, 4),
    index=["A", "B", "C", "D"],
    columns=["W", "X", "Y", "Z"]
)
print("DataFrame gốc:\n", df)

# 2) loc: hàng A,C và cột W,Y
loc_selection = df.loc[["A", "C"], ["W", "Y"]]
print("\nLấy bằng loc:\n", loc_selection)

# 3) iloc: 2 hàng cuối (2,3) và 2 cột đầu (0,1)
iloc_selection = df.iloc[2:4, 0:2]
print("\nLấy bằng iloc:\n", iloc_selection)

# 4) Giá trị tại hàng B, cột Z
value_BZ = df.loc["B", "Z"]
print("\nGiá trị tại [B, Z]:", value_BZ)

print("\nGiải thích: loc theo nhãn, iloc theo vị trí số nguyên.")
```

## Bài 2.4 - Reindex và điền giá trị thiếu

```python
import pandas as pd

# 1) Tạo Series ban đầu
product_stock = pd.Series([10, 50, 30], index=[0, 3, 5])
print("Series ban đầu:\n", product_stock)

# 2) Reindex từ 0 đến 5
reindexed = product_stock.reindex(range(6))
print("\nSeries sau reindex (có NaN):\n", reindexed)

# 3) Reindex + ffill
reindexed_ffill = product_stock.reindex(range(6), method="ffill")
print("\nSeries sau reindex + ffill:\n", reindexed_ffill)

print("\nGiải thích:")
print("- reindex thêm các chỉ số mới, vị trí chưa có dữ liệu sẽ là NaN.")
print("- ffill điền giá trị gần nhất phía trước.")
```

## Bài 2.5 - Sắp xếp, tính tổng điểm, thống kê mô tả

```python
import pandas as pd

# 1) Tạo bảng điểm
scores_data = {
    "Toan": [7, 5, 9, 8],
    "Ly": [8, 9, 6, 7],
    "Hoa": [6, 8, 9, 5]
}
student_scores = pd.DataFrame(scores_data, index=["Hùng", "Lan", "Minh", "An"])
print("Bảng điểm ban đầu:\n", student_scores)

# 2) Sắp xếp theo Toán giảm dần
sorted_scores = student_scores.sort_values(by="Toan", ascending=False)
print("\nBảng điểm sau khi sắp xếp theo Toán giảm dần:\n", sorted_scores)

# 3) Thêm cột tổng điểm
student_scores["Tong_diem"] = student_scores[["Toan", "Ly", "Hoa"]].sum(axis=1)
print("\nBảng điểm có cột tổng điểm:\n", student_scores)

# 4) Thống kê mô tả
desc = student_scores[["Toan", "Ly", "Hoa"]].describe()
print("\nThống kê mô tả:\n", desc)

print("\nGiải thích: describe() cho các thống kê quan trọng như mean, std, min, max, tứ phân vị.")
```
