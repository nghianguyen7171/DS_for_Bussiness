## Nguyên tắc cơ bản
<hr>

- Rõ ràng: loại bỏ yếu tố gây nhiễu (3D không cần thiết, hiệu ứng rườm rà, tỷ lệ trục méo), dùng mã hóa phù hợp (vị trí/độ dài tốt hơn góc/diện tích, màu sắc dùng tiết kiệm để phân nhóm hoặc nhấn mạnh).

- Trung thực: không “bẻ” trục gây hiểu lầm, không cắt bớt bối cảnh cần thiết, ghi đơn vị và nguồn dữ liệu đầy đủ.

- Có mục tiêu: mỗi biểu đồ nên trả lời một câu hỏi cụ thể; tiêu đề như một thông điệp thay vì mô tả máy móc. Như Wilke viết: “Good visualizations show the data clearly. Bad visualizations obscure it, or worse, mislead.” Hãy ưu tiên nội dung và khả dụng, rồi mới đến thẩm mỹ.

**✨ Một biểu đồ tốt có thể cô đọng câu chuyện dữ liệu:**

Mà hàng trăm dòng chữ khó truyền đạt được qua:

**📋 Các thành phần của câu chuyện dữ liệu:**
- **What/When/Where**: Mở ra bối cảnh
- **So what**: Nêu phát hiện chính
- **Now what**: Gợi ý hành động

**🎯 Ứng dụng trong các lĩnh vực:**

**🔬 Trong khoa học:**
- Minh họa giả thuyết và kết quả thí nghiệm

**💼 Trong kinh doanh:**
- Làm rõ hiệu quả, nguyên nhân gốc rễ
- Lựa chọn chiến lược

**📰 Trong truyền thông:**
- Giúp công chúng nắm bắt vấn đề phức tạp nhanh và đúng

**💡 Kỹ thuật tạo biểu đồ thuyết phục:**
- ✅ Dùng tiêu đề mang thông điệp
- ✅ Chú thích/annotation để dẫn mắt người xem
- ✅ Điểm nhấn bằng màu tiết kiệm để làm nổi bật điều quan trọng nhất

**📝 Nguyên tắc:**
Câu chuyện thuyết phục = **đúng dữ liệu** + **đúng bối cảnh** + **đúng khán giả**

# Nguyên tắc thiết kế tốt
<hr>

- Thiết kế tốt = dữ liệu rõ ràng, người xem hiểu nhanh, không bị lừa bởi hình thức.

- Thiết kế xấu = dữ liệu bị che giấu, gây nhầm lẫn, hoặc làm người xem hiểu sai.

- Trích dẫn: “Good design makes data clear. Bad design hides or misleads.” (Claus O. Wilke)

## Sử dụng màu sắc đúng cách
<hr>

Nguyên tắc cơ bản:

- Dữ liệu phân loại (categorical) → dùng bảng màu khác biệt (qualitative palette).

- Dữ liệu liên tục (sequential) → dùng thang màu từ nhạt → đậm.

- Dữ liệu có hai cực (diverging) → dùng thang đối xứng, ví dụ xanh – trắng – đỏ.

Tránh:

- Quá nhiều màu khiến người xem bị rối.

- Dùng màu không thân thiện với người mù màu (colorblind).

## Phân bố dữ liệu (Distributions)
<hr>

Histogram
- Dùng để hiển thị phân bố của dữ liệu liên tục.

- Dữ liệu được chia thành các “bin” (khoảng giá trị), mỗi cột biểu diễn tần suất.

- Quan trọng: chọn bin width hợp lý. Bin quá rộng → mất chi tiết; quá hẹp → nhiễu.

- Cho thấy hình dạng tổng thể của phân bố (chuẩn, lệch, nhiều đỉnh...).
![](https://raw.githubusercontent.com/toantran249/NEU_KHDL/6c436952736e736ff2357efdeac9d3406dcf27ff/12_slide04_histogram.png)

Density plot
- Là phiên bản “mềm” của histogram, ước lượng mật độ xác suất.

- Giúp hình dung phân bố mượt hơn, không phụ thuộc vào số bin.

- Rất trực quan khi so sánh nhiều phân bố (ví dụ nam vs nữ).

- Tuy nhiên cần giải thích cẩn thận vì trục y là mật độ, không phải số lượng tuyệt đối.
![](https://raw.githubusercontent.com/toantran249/NEU_KHDL/6c436952736e736ff2357efdeac9d3406dcf27ff/12_slide05_density.png)

## Điều chỉnh khoảng cách xung quanh các subplot
<hr>

**📐 Điều chỉnh khoảng cách giữa các subplot:**
- `matplotlib` để lại một khoảng đệm nhất định xung quanh bên ngoài các subplot
- Khoảng cách giữa các subplot cũng được chỉ định
- Được chỉ định tương đối với chiều cao và chiều rộng của đồ thị
- Nếu bạn thay đổi kích thước đồ thị, đồ thị sẽ tự động điều chỉnh
- Bạn có thể thay đổi khoảng cách bằng phương thức `subplots_adjust`
- Cho phép kiểm soát chính xác vị trí và khoảng cách của các subplot

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.standard_normal(500), bins=50, color="black", alpha=0.5)
fig.subplots_adjust(wspace=0, hspace=0)
plt.savefig("chart5.png")
```

## Chú thích vẽ trên một đồ thị con
<hr>

**✏️ Chú thích và văn bản trên đồ thị:**
- Vẽ các chú thích đồ thị tùy chỉnh: văn bản, mũi tên hoặc các hình dạng khác
- Sử dụng các hàm `text`, `arrow`, và `annotate`

**📝 Hàm text:**
- `ax.text(x, y, "Hello world!", family="monospace", fontsize=10)`
- Vẽ văn bản tại các tọa độ (x, y) đã cho trên đồ thị với kiểu tùy chỉnh

**🔗 Hàm annotate:**
- Chú thích có thể bao gồm cả văn bản và mũi tên chỉ vào vị trí cụ thể

```python
from datetime import datetime
# Create spx.csv file for the example
spx_csv_content = "date,SPX\n2003-01-02,909.03\n2007-05-04,1499.26\n2007-10-11,1554.41\n2008-03-12,1290.35\n2008-09-15,1172.22\n2010-12-31,1257.64\n2011-01-03,1271.87"
with open("spx.csv", "w") as f:
    f.write(spx_csv_content)

fig, ax = plt.subplots(figsize=(12, 6)) # Tăng kích thước để dễ nhìn hơn
data = pd.read_csv("spx.csv", index_col=0, parse_dates=True)
spx = data["SPX"]

spx.plot(ax=ax, color="black")

crisis_data = [
    (datetime(2007, 10, 11), "Peak of bull market"),
    (datetime(2008, 3, 12), "Bear Stearns Fails"),
    (datetime(2008, 9, 15), "Lehman Bankruptcy")
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor="black", headwidth=4, width=2,
                                headlength=4),
                horizontalalignment="left",
                verticalalignment="top")

# Zoom in on 2007-2010
ax.set_xlim(["1/1/2007", "1/1/2011"])
ax.set_ylim([600, 1800])
ax.set_title("Important dates in the 2008–2009 financial crisis")
plt.savefig("chart13.png")
```

![](https://raw.githubusercontent.com/toantran249/NEU_KHDL/70c3dcccbc152116fd201f1b380767d116132165/chart13.png)

**📍 Điểm quan trọng trong chú thích:**
- `ax.annotate`: vẽ các nhãn tại các tọa độ x và y được chỉ định
- `set_xlim` và `set_ylim`: đặt thủ công các ranh giới bắt đầu và kết thúc cho đồ thị
- Thay vì sử dụng mặc định của matplotlib
- `ax.set_title`: thêm một tiêu đề chính cho đồ thị

**📚 Tài liệu:**
Xem thư viện matplotlib để biết thêm các tùy chọn định dạng văn bản chi tiết

**🔷 Vẽ các hình dạng:**
- Việc vẽ các hình dạng đòi hỏi sự cẩn thận và tỉ mỉ hơn
- `matplotlib` có các đối tượng đại diện cho nhiều hình dạng phổ biến, gọi là `patches`
- Một số như `Rectangle` và `Circle` được tìm thấy trong `matplotlib.pyplot`
- Thông tin đầy đủ nằm trong `matplotlib.patches`

**✅ Cách sử dụng:**
- Tạo đối tượng `patches` và thêm nó vào một đồ thị con `ax` bằng phương thức `add_patch`

```python
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color="black", alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color="blue", alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color="green", alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
plt.savefig("chart14.png")
```

## Cấu hình matplotlib
<hr>

`matplotlib` đi kèm với các lược đồ màu và các giá trị mặc định hướng đến việc chuẩn bị cho các hình sẵn sàng lưu vào máy. Hầu như tất cả các tham số mặc định đều có thể được tùy chỉnh thông qua các tham số toàn cục quy định kích thước hình, khoảng cách subplot, màu sắc, kích thước phông chữ, kiểu lưới, v.v. Một cách để sửa đổi cấu hình theo chương trình từ Python là sử dụng phương thức `rc`; ví dụ, để đặt kích thước hình mặc định toàn cục là 10 × 10, bạn có thể nhập:

Đối số đầu tiên cho `rc` là thành phần bạn muốn tùy chỉnh, chẳng hạn như `"figure"`, `"axes"`, `"xtick"`, `"ytick"`, `"grid"`, `"legend"`, hoặc nhiều thành phần khác. Để tùy chỉnh rộng hơn và xem danh sách tất cả các tùy chọn, matplotlib đi kèm với một tệp cấu hình `matplotlibrc` trong thư mục `matplotlib/mpl-data`. Nếu bạn tùy chỉnh tệp này và đặt nó trong thư mục chính của mình với tên `.matplotlibrc`, nó sẽ được tải mỗi khi bạn sử dụng matplotlib.

Như chúng ta sẽ thấy trong phần tiếp theo, thư viện `seaborn` có một số chủ đề hoặc kiểu đồ thị tích hợp sẵn sử dụng hệ thống cấu hình của matplotlib bên trong.

# Vẽ đồ thị với pandas và seaborn
<hr>

`matplotlib` là một công cụ trực quan hóa dữ liệu ở mức độ khá cơ bản. Giống như bạn đang *lắp ráp* một đồ thị từ các thành phần cơ bản của nó bao gồm: hiển thị dữ liệu, chú giải, tiêu đề, nhãn vạch chia và các chú thích khác.

Trong `pandas`, chúng ta có thể có nhiều cột dữ liệu, cùng với các chỉ số tên hàng và tên cột. Bản thân `pandas` có các phương thức tích hợp sẵn giúp đơn giản hóa việc tạo hình trực quan từ các đối tượng `DataFrame` và `Series`. Một thư viện khác là `seaborn`, một thư viện đồ họa thống kê cấp cao được xây dựng trên `matplotlib`. Ưu điểm của seaborn là đơn giản hóa việc tạo nhiều loại hình trực quan phổ biến.

```python
import warnings
warnings.filterwarnings('ignore')

# Bảo đảm hiển thị inline an toàn trên nhiều môi trường
try:
    get_ipython().run_line_magic('matplotlib', 'inline')
except Exception:
    pass

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Tạo DataFrame tổng hợp mô phỏng doanh thu theo tháng cho 2 sản phẩm
np.random.seed(0)
months = pd.date_range('2024-01-01', periods=12, freq='M')
df_demo = pd.DataFrame({
    'Month': months,
    'Revenue_A': np.random.randint(80, 140, size=12).cumsum(),
    'Revenue_B': np.random.randint(60, 120, size=12).cumsum()
})
df_demo['Total'] = df_demo['Revenue_A'] + df_demo['Revenue_B']

print('✅ Tạo DataFrame tổng hợp thành công!')
df_demo.head()

# 1) pandas.plot: vẽ đường thời gian cho hai sản phẩm
ax = df_demo.set_index('Month')[['Revenue_A', 'Revenue_B']].plot(
    figsize=(8, 4), linewidth=2, marker='o', title='Doanh thu theo tháng (pandas.plot)'
)
ax.set_xlabel('Tháng')
ax.set_ylabel('Doanh thu (đv giả lập)')
plt.tight_layout()
plt.savefig('chart_pandas_demo.png')
plt.show()

# 2) seaborn: chuyển sang dạng long và vẽ barplot theo tháng
long_df = df_demo.melt(id_vars='Month', value_vars=['Revenue_A', 'Revenue_B'],
                       var_name='Product', value_name='Revenue')

plt.figure(figsize=(9, 4))
sns.barplot(data=long_df, x='Month', y='Revenue', hue='Product', alpha=0.9)
plt.title('So sánh doanh thu theo tháng (seaborn.barplot)')
plt.xlabel('Tháng')
plt.ylabel('Doanh thu (đv giả lập)')
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig('chart_seaborn_bar_demo.png')
plt.show()

# 3) seaborn.lineplot: đường xu hướng tổng doanh thu
plt.figure(figsize=(8, 4))
sns.lineplot(data=df_demo, x='Month', y='Total', marker='o', linewidth=2, color='#2563eb')
plt.title('Xu hướng tổng doanh thu (seaborn.lineplot)')
plt.xlabel('Tháng')
plt.ylabel('Tổng doanh thu (đv giả lập)')
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig('chart_seaborn_line_demo.png')
plt.show()
```

Hãy xem một ví dụ về tập dữ liệu tiền boa nhà hàng. Giả sử chúng ta muốn tạo một đồ thị cột xếp chồng hiển thị tỷ lệ phần trăm các điểm dữ liệu cho mỗi quy mô bữa tiệc cho mỗi ngày. Chúng ta tải dữ liệu bằng `read_csv` và tạo một bảng chéo theo ngày và quy mô bữa tiệc. Hàm `pandas.crosstab` là một cách thuận tiện để tính toán một bảng tần suất đơn giản từ hai cột DataFrame:

```python
# Dữ liệu mô phỏng thực tế với các mẫu về tip percentage theo ngày và thời gian
days = ['Thur', 'Fri', 'Sat', 'Sun']
times = ['Lunch', 'Dinner']
sexes = ['Male', 'Female']
smokers = ['Yes', 'No']

# Tạo ~200 observations để có đủ dữ liệu cho các biểu đồ
tips_data = []
np.random.seed(42)  # Để có thể tái lập kết quả

for _ in range(200):
    # Chọn ngày (cuối tuần có nhiều khách hơn)
    day = np.random.choice(days, p=[0.15, 0.15, 0.35, 0.35])  # Sat, Sun nhiều hơn

    # Chọn thời gian (dựa vào ngày)
    if day in ['Sat', 'Sun']:
        time = np.random.choice(times, p=[0.1, 0.9])  # Cuối tuần chủ yếu Dinner
    else:
        time = np.random.choice(times, p=[0.6, 0.4])  # Ngày thường có cả Lunch

    # Quy mô bữa tiệc (size) - từ 1 đến 6 người
    size = np.random.choice([1, 2, 3, 4, 5, 6], p=[0.05, 0.35, 0.25, 0.20, 0.10, 0.05])

    # Tổng hóa đơn - phụ thuộc vào size và time
    base_bill = 15 + size * 8  # Bill cơ bản
    if time == 'Dinner':
        base_bill *= 1.3  # Dinner thường đắt hơn
    if day in ['Sat', 'Sun']:
        base_bill *= 1.1  # Cuối tuần có thể đắt hơn
    total_bill = np.random.normal(base_bill, base_bill * 0.3)
    total_bill = max(5, total_bill)  # Đảm bảo tối thiểu $5

    # Tính tip - phụ thuộc vào nhiều yếu tố
    # Tip percentage thường 15-20%, nhưng có thể thay đổi
    base_tip_pct = 0.17
    if day in ['Sat', 'Sun']:
        base_tip_pct += 0.02  # Cuối tuần tip tốt hơn
    if size >= 4:
        base_tip_pct -= 0.01  # Nhóm lớn tip ít hơn một chút
    tip_pct = np.random.normal(base_tip_pct, 0.05)
    tip_pct = max(0.05, min(0.30, tip_pct))  # Giới hạn từ 5% đến 30%

    tip = total_bill * tip_pct

    # Các thuộc tính khác
    sex = np.random.choice(sexes)
    smoker = np.random.choice(smokers, p=[0.38, 0.62])  # Khoảng 38% hút thuốc

    tips_data.append({
        'total_bill': total_bill,
        'tip': tip,
        'sex': sex,
        'smoker': smoker,
        'day': day,
        'time': time,
        'size': size
    })

tips = pd.DataFrame(tips_data)

# Tính tip_pct (tip percentage) - sẽ được dùng trong các ví dụ
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])

print(f"\n✅ Tips DataFrame đã được tạo thành công!")
print(f"   - Số hàng: {len(tips)}")
print(f"   - Các cột: {', '.join(tips.columns.tolist())}")
print(f"\n📋 5 hàng đầu tiên của tips DataFrame:")
print(tips.head())
```

Các điểm dữ liệu được chia vào các miền rời rạc, cách đều nhau và số lượng điểm dữ liệu trong mỗi miền được vẽ bằng 1 thanh có chiều cao tương đương với số lượng điểm. Một loại đồ thị liên quan là đồ thị hàm mật độ, được hình thành bằng cách tính toán ước tính của một phân phối xác suất liên tục có thể đã tạo ra dữ liệu quan sát được. Quy trình thông thường là xấp xỉ phân phối này dưới dạng hỗn hợp (mixed) của các kernel - là các phân phối đơn giản hơn như phân phối chuẩn. Sử dụng `plot.density` tạo một đồ thị mật độ bằng cách sử dụng ước tính hỗn hợp các phân phối chuẩn thông thường như sau:

## Lưới đa điểm và dữ liệu phân loại
<hr>

Một cách để trực quan hóa dữ liệu với nhiều biến phân loại là sử dụng lưới đa diện - một bố cục hai chiều của các đồ thị trong đó dữ liệu được chia trên các đồ thị trên mỗi trục dựa trên các giá trị riêng biệt của một biến nhất định. `seaborn` có một hàm tích hợp hữu ích `catplot` giúp đơn giản hóa việc tạo nhiều loại đồ thị đa diện được chia theo các biến phân loại:

```python

tips["tip_pct"] = tips["tip"] / (tips["total_bill"] - tips["tip"])
sns.catplot(x="day", y="tip_pct", hue="time", col="smoker",
            kind="bar", data=tips[tips.tip_pct < 1])
plt.savefig("chart29.png")
```
